```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("train.csv")
```

```python
# Check data types, null values
print(df.info())
```

```python
# Statistical summary
print(df.describe())
```

```python
# Value counts for categorical features
print(df['Survived'].value_counts())
print(df['Pclass'].value_counts())
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', palette='Set2')
plt.show()
```

```python
df['Age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='Set1')
plt.title('Age vs Fare (Colored by Survival)')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')  
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', hue='Survived', data=df, palette='Set3', legend=False)
plt.title('Age vs Survival')
plt.show()

sns.boxplot(x='Survived', y='Age', hue='Survived', data=df, palette='Set3', legend=False)
plt.title('Fare vs Survival')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Survived', data=df, palette='pastel', legend=False)
plt.title('Survival Count')
plt.show()
```

```python
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', hue='Sex', data=df, palette='muted', legend=False)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', hue='Pclass', data=df, palette='cool', legend=False)
plt.title('Survival Rate by Passenger Class')
plt.show()
```
